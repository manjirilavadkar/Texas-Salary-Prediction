#### 'HRS PER WK','MONTHLY','HRLY RATE','ANNUAL' columns have outliers

```python
num_col=data[['HRS PER WK','MONTHLY','ANNUAL']]
for col in num_col:
    Q1=data[col].quantile(0.25)
    Q3=data[col].quantile(0.75)
    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR
    print("***********")
    print("Column Name:",col)
    print("IQR  ",IQR)
    print("Lower  ",lower)
    print("Upper  ",upper)
    print("Outliers Percentage",len(data.loc[(data[col]<lower)|(data[col]>upper)])/len(data))
    print("***********")
```

# Task 3.2:- What departments/roles have the biggest wage disparities between managers and employees?

```python
#Load data
pd.set_option('display.max_columns', None)
new_data=pd.read_csv("salary.csv")
new_data
```

```python
# drop irrlevant columns
```

```python
new_data.drop(['duplicated','multiple_full_time_jobs','combined_multiple_jobs','summed_annual_salary','hide_from_search'],axis=1,inplace=True)
```

```python
new_data.rename(columns={'AGENCY NAME':'AGENCY_NAME'},inplace=True)
```

```python
#New column which divides job roles in to categories manager and employee
```

```python
job_role=[]

for i in new_data['CLASS TITLE']:
    if 'MANAGER' in i or 'MGR' in i:
        job_role.append("MANAGER")
    else:
        job_role.append("EMPLOYEE")
new_data['MNG_VS_EMP']=job_role
```

```python
new_data.loc[new_data['MNG_VS_EMP'].str.contains('MANAGER')]
```

```python

import scipy.stats as stats
import statsmodels.api as sm

pivot_table = new_data.pivot_table(index='AGENCY_NAME', columns='MNG_VS_EMP', values='ANNUAL', aggfunc='mean')
pivot_table
```

```python
pivot_table.fillna(0,inplace=True)
```

```python
pivot_table
```

```python
pivot_table['diff_sal']=pivot_table['MANAGER']-pivot_table['EMPLOYEE']
```

#### Top 10 departments which has biggest disparity between manager and employee salaries

```python
pivot_table['diff_sal'].nlargest(n=10)
```

```python
plt.figure(figsize=(25,10))
pivot_table.drop('diff_sal',axis=1).plot(kind='bar',figsize=(22,7))
plt.xlabel('Department')
plt.ylabel('Mean Salary')
plt.title('Mean Salary by Department and Job Role')
plt.legend(title='Job Role')
plt.xticks(rotation=90)

plt.show()
```

### There is a significant disparity in salaries of the Manager and Employee in all departments.
### In most of the Departments Manager has higher salary than other job roles

# Task 3.3:- Have salaries and total compensations for some roles/ departments/ head-count changed over time

```python
#Load data
pd.set_option('display.max_columns', None)
new_data=pd.read_csv("salary.csv")
new_data.head()
```

```python
new_data.drop(['duplicated','multiple_full_time_jobs','combined_multiple_jobs','summed_annual_salary','hide_from_search'],axis=1,inplace=True)
```

```python
top_agencies=new_data['AGENCY NAME'].value_counts().nlargest(10)
```

```python
top_agencies=top_agencies.index.to_list()
top_agencies
```

```python
new_data=new_data.loc[new_data['AGENCY NAME'].isin(top_agencies)]
```

```python
new_data['EMPLOY DATE'] = pd.to_datetime(new_data['EMPLOY DATE'])
year=new_data['EMPLOY DATE'].dt.year
```

```python
pivot_tab = new_data.pivot_table(index=year,columns='AGENCY NAME', values='ANNUAL',aggfunc='mean')
pivot_tab
```

```python
pivot_tab.drop([2069,2071,2072],axis=0,inplace=True)
```

```python
pivot_tab
```

### Plots of 10 department which have large number of records of employees

```python

plt.figure(figsize=(20,25))
num=1
for col,name in zip(pivot_tab,top_agencies):
    if num<=10:
        plt.subplot(5,2,num)
        plt.plot(pivot_tab[col])
        pivot_tab[col].dropna(axis=0)
        plt.xlabel('Year')
        plt.ylabel('ANNUAL')
        plt.title('Departments Name= '+name)
    num+=1
```


### Salaries of largest 10 departments varies over time
